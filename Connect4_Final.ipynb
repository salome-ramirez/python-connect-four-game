```python
# Name: Salome Collante-Ramirez Spring 2025


#----------Grid Size-------
count_row = 6
count_col = 7

#----------Game Board-------
def new_gameboard():
  gameboard= []
  for i in range(count_row):
    row =[]

    for j in range (count_col):
      row.append('.')
    gameboard.append(row)
  return gameboard


def showboard(gameboard):
  for row in gameboard:
    print(" ".join(row))
  print("1 2 3 4 5 6 7 ")

#dropping tokens into board
def drop(token,colnum, gameboard):
  for row in gameboard[::-1]:
      if row[colnum-1]=='.':
        row[colnum-1]=token
        return True
  return False

#------Winning Conditions--------
def four_in_a_row(gameboard):
  #horizontally
  for row in gameboard:
    for index in range(4):
      if row[index] != '.':
        if row[index] == row[index+1] and \
        row[index] == row[index+2] and \
        row[index] == row[index+3]:
          #capitilize winning token
          for i in range(4):
            row[index + i] = row[index + i].upper()
          return True

  #vertically
  for col_index in range(7):
    for row_index in range(3):
      if gameboard[row_index][col_index] != '.':
        if gameboard[row_index][col_index] == gameboard[row_index+1][col_index] and \
        gameboard[row_index][col_index] == gameboard[row_index+2][col_index] and \
        gameboard[row_index][col_index] == gameboard[row_index+3][col_index]:
          for i in range(4):
              gameboard[row_index + i][col_index]= gameboard[row_index + i][col_index].upper()
          return True



  #diagonally (R to L) row range (4) - col range (3)
  for row_index in range(3):
    for col_index in range(4):
      if gameboard[row_index][col_index] != '.':
        if gameboard[row_index][col_index] == gameboard[row_index+1][col_index+1] and \
        gameboard[row_index][col_index] == gameboard[row_index+2][col_index+2] and \
        gameboard[row_index][col_index] == gameboard[row_index+3][col_index+3]:
          for i in range(4):
              gameboard[row_index + i][col_index+i]= gameboard[row_index + i][col_index+i].upper()
          return True
   #diagonally L to R --- row range (4) and col range (3,7)
  for row_index in range(3):
      for col_index in range(3,7):
        if gameboard[row_index][col_index] != '.':
          if gameboard[row_index][col_index] == gameboard[row_index+1][col_index-1] and \
          gameboard[row_index][col_index] == gameboard[row_index+2][col_index-2] and \
          gameboard[row_index][col_index] == gameboard[row_index+3][col_index-3]:
            for i in range(4):
              gameboard[row_index + i][col_index-i]= gameboard[row_index + i][col_index-i].upper()
            return True
  return False
#---------Draw-----------
def draw(gameboard):
  for row in gameboard:
    for column in row:
      if column== '.':
        return False
    return True


#--------Game loop-----------------------------------
def play_game():
  gameboard = new_gameboard()
  showboard(gameboard)
  current_player = 'x'
  while True:
    try:
      play_column= int(input(f'Player {current_player} - Enter desired column 1-7:  '))
    except ValueError:
      print('Not quite...input must be an integer from 1-7')
      continue
    if 1 > play_column or play_column >7:
      print('Column selection must be 1-7')
      continue

    drop(current_player,play_column,gameboard)
    showboard(gameboard)
    if draw(gameboard):
      print("It's a draw!!!")
      game_over = True
    elif four_in_a_row(gameboard):
      print(f'Connect Four! {current_player} wins!')
      showboard(gameboard)
      game_over = True
    else:
      if current_player == 'x':

        current_player = 'o'
      else:
        current_player = 'x'
      game_over = False
#---------play again?-----------
    if game_over:
      play_again=None
      while play_again != 'y' and play_again != 'n':
        play_again=input('Play again? y/n ')
      if play_again == 'y':
        play_game()
      else:
        print('Goodbye!')
        break



play_game()

```
